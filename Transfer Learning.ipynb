```python
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import Dense, Conv2D, Input
from tensorflow.keras.models import Model
```

```python
mobile_net=MobileNet()

#inputLayer
mobile_net.summary()  #Pailei use vako file user garera model banauxau. Pailei train gareko model linxau
```

```python
mobile_net.summary()
```

```python
x= mobile_net.output #0pading le edge haru ma 0 halxa
x=Dense(2, activation='softmax')(x) #yo hamro layer. yeslai lagera mathi ko layer ma jodeko. So neuron ma 1000*2 kota hunxa
model=Model(inputs=mobile_net.input, outputs=x)
```

```python
model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
trainDatagen = ImageDataGenerator(rescale=1./255)

testDatagen = ImageDataGenerator(rescale=
                                 1./255)
validDatagen = ImageDataGenerator(rescale=1./255)
s=224 #Shape of our image

train_set=trainDatagen.flow_from_directory('data/train',target_size=(s,s),color_mode='rgb',batch_size=32, class_mode='categorical')
```

```python
test_set=testDatagen.flow_from_directory('data/test',target_size=(s,s),color_mode='rgb',batch_size=32, class_mode='categorical')
```

```python
valid_set=validDatagen.flow_from_directory('data/valid',target_size=(s,s),color_mode='rgb',batch_size=32, class_mode='categorical')
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])
```

```python
history=model.fit_generator(train_set,steps_per_epoch=10,epochs=2,validation_data=valid_set,validation_steps=10)
```

```python
score=model.evaluate_generator(test_set,steps=1)
print(f"Test loss:{score[0]} Test Accuracy:{score[1]}")
```

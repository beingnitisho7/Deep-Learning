```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/coin.png', 1)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB) #changes two chanels.. RGB ko
plt.imshow(image)
```

```python
resized = cv2.resize(image, None, fx=0.5, fy=0.5)
plt.imshow(resized)
```


# Image Threshold

```python
_,thresholed=cv2.threshold(image,100,100,240, cv2.THRESH_BINARY_INV) #Thresholding le certain value ma lanxa
plt.imshow(thresholed, cmap='gray')
```

# Image Masking

```python
img_Ev = cv2.imread('images/everest.jpg', 1)
```

```python
img_flo=cv2.imread('images/petal.jpg', 1)
```

```python
plt.imshow(img_Ev, cmap='gray')
plt.imshow(img_flo, cmap='gray')
```

```python
backg = cv2.cvtColor(img_Ev, cv2.COLOR_BGR2GRAY)
plt.imshow(backg)
plt.title('GRAY image')
plt.show()
print(backg.shape)
```

```python
forg = cv2.cvtColor(img_flo, cv2.COLOR_BGR2GRAY)
plt.imshow(forg)
plt.title('Gray image')
plt.show()
print(forg.shape)
```

```python
lv=np.array([0,0,0])

hv=np.array([220,220,220])

mask=cv2.inRange(img_flo,lv,hv)
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
masked_img=img_flo.copy()
masked_img[mask!=255]=[0,0,0] #Sabai lai kaalo banaunu xa
plt.imshow(masked_img)
plt.show()
print(masked_img.shape)
```

```python
resized_bag=cv2.resize(masked_img, (3157,2104))
```

```python
plt.imshow(resized_bag)
```
